```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("BigBazaar (1).csv").dropna().drop_duplicates()
```

```python
data.shape
```

```python
data.sample(5)
```

```python
data["Category"].unique()
```

```python
categories_to_exclude = ["Home Appliances", "Meats & Seafoods","Electronics","Toys, Sports & Games","Luggage","Bathware & Organizers",'Home & Kitchen Appliances','Kitchen Appliances',"Home Fashion","Women","Footwear"]
filtered_data = data[~data['Category'].isin(categories_to_exclude)]
```

```python
filtered_data.shape
```

```python
columns_to_drop = ['DiscountedPrice', 'Category', 'SubCategory','Description',"LongDescription"]  # Replace with actual column names
filtered_data = filtered_data.drop(columns=columns_to_drop)
```

```python
filtered_data = filtered_data.reset_index(drop=True)
```

```python
filtered_data
```

```python
unique_product_names = filtered_data['Name'].unique()

sales_data = {'product_names': unique_product_names}
sales_df = pd.DataFrame(sales_data)

sales_df['Last_30_weeks_sales'] = [list(np.random.randint(10, 101, size=30)) for _ in range(len(unique_product_names))]
```

```python
sales_df['inventory'] = [
    [sale + np.random.randint(1, 101) for sale in weekly_sales]
    for weekly_sales in sales_df['Last_30_weeks_sales']
]
```

```python
sales_df["Selling_Price"]=filtered_data["Price"]
```

```python
sales_df['Selling_Price'] = sales_df['Selling_Price'].replace(',', '', regex=True).astype(float)
```

```python
wholesale_price_difference = np.random.uniform(0, 0.15, size=(len(sales_df), 3))

sales_df['Supplier_A'] = sales_df['Selling_Price'] * (1 - wholesale_price_difference[:, 0])
sales_df['Supplier_B'] = sales_df['Selling_Price'] * (1 - wholesale_price_difference[:, 1])
sales_df['Supplier_C'] = sales_df['Selling_Price'] * (1 - wholesale_price_difference[:, 2])
```

```python
sales_df['Supplier_A'] = sales_df['Supplier_A'].round(0)
sales_df['Supplier_B'] = sales_df['Supplier_B'].round(0)
sales_df['Supplier_C'] = sales_df['Supplier_C'].round(0)
```

```python
sales_df
```

```python
sales_df.to_csv('processed_dataset.csv', index=False)
```
